````python
#Drop the column which has most of null values
df_Airbnb.drop(['last_review','reviews_per_month'],axis=1,inplace=True)
````

````python
#Replace null values to not available
null_values_host_name= list(df_Airbnb[pd.isnull(df_Airbnb['host_name'])].index.values)
df_Airbnb.iloc[null_values_host_name,[3]]='NOT AVAILABLE'
````

````python
#Drop the name column
df_Airbnb.drop(['name'],axis=1,inplace=True)
````

````python
#Check the null values
null_values_in_our_data=df_Airbnb.isnull().sum().sort_values(ascending=False)
print(null_values_in_our_data)
````

````python
df_Airbnb.describe()
````

````python
#Check the rows whose price is zero
df_Airbnb[df_Airbnb['price']==0]
````

````python
#Mean value of price
mean_value_of_price=df_Airbnb['price'].mean()
````

````python
#Replace to mean value of the price from zero
price_column=df_Airbnb['price']
price_column.replace(to_replace=0,value=mean_value_of_price,inplace=True)
````

````python
df_Airbnb.describe()
````

**1.Difference of traffic among different area** 

````python
#Analysing the traffic among area
plt.figure(figsize=(10,8))
sns.scatterplot(x=df_Airbnb['longitude'],y=df_Airbnb['latitude'],hue=df_Airbnb['neighbourhood_group']).set_title('Density of Hotels')
````

**2.How many hotels in the neighbourhood group**

````python
#Number of hotels in the neighbourhood group
hotels_in_neighbourhood_group=df_Airbnb.groupby(['neighbourhood_group'])['id'].count().reset_index()
hotels_in_neighbourhood_group.rename({'id':'no_of_hotels'},inplace=True,axis='columns')
print(hotels_in_neighbourhood_group)
````

````python
plt.figure(figsize=(12,6))
plt.plot(hotels_in_neighbourhood_group.neighbourhood_group,hotels_in_neighbourhood_group.no_of_hotels,marker='s',c='r',ls='-');
````

***The highest number of hotels in 'Manhattan' neighbourhood group while 'Staten Island' has the least.***

**3.Determine the count of neighbourhood group**

````python
#Count of neighbourhood group
neighbourhood_group=df_Airbnb['neighbourhood_group'].value_counts()
neighbourhood_group.plot(kind='pie',figsize=(12,6),autopct='%1.2f%%')
````

***'Manhattan' has the highest count of neighbourhood group while 'Staten Island' has the least.***

**4.Determine most popular neighbourhood**

````python
#Top 20 popular neighbourhood 
neighbourhood=df_Airbnb['neighbourhood'].value_counts().sort_values(ascending=False)[0:20]
neighbourhood.plot(kind='bar',figsize=(8,4))
````

***The williamsburg is the most popular neighbourhood***

**5.What is the average price of different room type according to different neighbourhood group**

````python
#Average price of room according to different neighbourhood group
price_vs_room_type=df_Airbnb.groupby(['neighbourhood_group','room_type'])['price'].mean().round(2).unstack()
print(price_vs_room_type)
````

````python
price_vs_room_type.plot(kind='bar',ylabel='price',title='average price of different type of room in every neighbourhood group',figsize=(6,6))
````

***'Entire home' has the highest average price in the  Manhattan group while 'Shared room' has the least***

**6.Which hosts are the  busiest and why**

````python
#Busiest hosts
busiest_hosts=df_Airbnb.groupby(['host_name','neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
busiest_hosts=busiest_hosts.sort_values(by='minimum_nights',ascending=False).head(5)
print(busiest_hosts)
````

````python
#Creating the bar plot
name=busiest_hosts['host_name']
stayed=busiest_hosts['minimum_nights']
fig=plt.figure(figsize=(12,6))
plt.bar(name,stayed,color='green',width=0.4)
plt.xlabel('name of the host')
plt.ylabel('minimum number of nights stayed')
plt.title('busiest hosts')
plt.show()
````

***According to the busiest host 'Sonder(NYC)' is the most busiest while 'David' is the least busiest***

**Availability:**

**7.Determine the availability of room at different neighbourhood group and average availability of room at different room type.**

````python
#availability of room in different neighbourhood group
plt.figure(figsize=(12,6))
plt.title('availability of room in different neighbourhood group',color='green',fontsize=14)
sns.boxplot(data=df_Airbnb,x='neighbourhood_group',y='availability_365',palette='dark')
plt.show()
````

**Here 'Manhattan' has highest availability of room while 'Staten Island' has least availability of room according to different neighbourhood group**

````python
#Average availability of room at different room type
plt.figure(figsize=(12,5))
plt.title('average availability of room at different room type',color='black',fontsize=16)
result_room=df_Airbnb.groupby(['room_type'])['availability_365'].aggregate(np.mean).reset_index().sort_values('availability_365')
sns.barplot(data=df_Airbnb,x='room_type',y='availability_365',order=result_room['room_type'])
plt.show()
````

**Here the highest average availability of room is 'Shared room' while 'Private room' or 'Entire home' has least average availability at different room type.**

**8.Determine the avrage of minimum night stayed in different neighbourhood group and room type.**

````python
#Average of minimum night stayed in different neighbourhood group
plt.figure(figsize=(12,8))
plt.title('average of minimum night stayed in different neighbourhood group',color='purple',fontsize=18)
result_group=df_Airbnb.groupby(['neighbourhood_group'])['minimum_nights'].aggregate(np.mean).reset_index().sort_values('minimum_nights')
sns.barplot(data=df_Airbnb,x='neighbourhood_group',y='minimum_nights',order=result_group['neighbourhood_group'])
plt.show()
````

**Here Manhattan has highest average of minimum night stayed in different neighbourhood group while bronx has least average of minimum night stayed**

````python
#Average night stayed at different room type
plt.figure(figsize=(12,8))
plt.title('average night stayed at different room type',color='black',fontsize=16)
result_room=df_Airbnb.groupby(['room_type'])['minimum_nights'].aggregate(np.mean).reset_index().sort_values('minimum_nights')
sns.barplot(data=df_Airbnb,x='room_type',y='minimum_nights',order=result_room['room_type'])
plt.show()
````

**From the above analysis, we can say that 'Entire home/apt' has highest average night stayed while 'private room' has least average night stayed at different room type.**

**9.Analysing the revenue in neighbourhood_group**

````python
#Fetching out the total cost
df_Airbnb['Total_cost']=df_Airbnb['price']*df_Airbnb['minimum_nights']

````

````python
#which neighbourhood_group will profit the most
df_Airbnb[['neighbourhood_group','Total_cost']].groupby(['neighbourhood_group']).sum('Total_cost').sort_values(by = 'Total_cost',ascending=False).head().plot(kind='barh')
````

***Manhattan has the highest revenue while staten Island has the least according to the neighbourhood group***
